````python
@setup
def main():
    n = ii()
    arrival = [0] * n
    leaving = [0] * n
    for i in range(n):
        arrival[i], leaving[i] = li()
    arrival.sort()
    leaving.sort()
    i = j = 0
    occupied = 0
    max_occupied = 0
    while i < n:
        if arrival[i] < leaving[j]:
            occupied += 1
            max_occupied = max(occupied, max_occupied)
            i += 1
        else:
            occupied -= 1
            j += 1
    
    print(max_occupied)
````
