```python
from collections import Counter
```

```python
@setup
def main():
    s = input().strip()
    freq = Counter(s)

    odd_chars = [ch for ch, cnt in freq.items() if cnt % 2 == 1]

    if len(odd_chars) > 1:
        print("NO SOLUTION")
    else:
        left = []
        mid = ""

        for ch in sorted(freq.keys()):
            cnt = freq[ch]
            if cnt % 2 == 1:
                mid = ch * cnt
            else:
                left.append(ch * (cnt // 2))
                
        left = "".join(left)
        right = left[::-1]

        print(left + mid + right)
```

```python
main()
```
