```python
@setup
def main():
    n = ii()
    arr = li()
    moves = 0
    for i in range(1, n):
        if arr[i] < arr[i-1]:
            moves += arr[i-1] - arr[i]
            arr[i] = arr[i-1]
    print(moves)
```
