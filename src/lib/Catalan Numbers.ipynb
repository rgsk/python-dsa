```python
def nthCatalan(n: int):
    if n == 0 or n == 1:
        return 1
    total = 0
    for i in range(n):
        total += nthCatalan(i) * nthCatalan(n - i - 1)
    return total
```

```python
def nthCatalan(n: int):
    if n == 0 or n == 1:
        return 1
    dp = [0 for _ in range(n + 1)]
    dp[0] = 1
    dp[1] = 1
    for j in range(2, n + 1):
        for i in range(j):
            dp[j] += dp[i] * dp[j-i-1]
    return dp[n]
```

```python
n = 3
nthCatalan(n)
```
