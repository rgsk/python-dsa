```python
from collections import defaultdict
from math import ceil
from typing import List


class Solution:
    def numRabbits(self, answers: List[int]) -> int:
        mp = defaultdict(int)
        for v in answers:
            mp[v] += 1
        result = 0
        for k in mp:
            seats = k+1
            result += ceil(mp[k]/seats) * seats
        return result
```

```python
import collections

class Solution:
    def numRabbits(self, answers):
        c = collections.Counter()
        res = 0
        for i in answers:
            if c[i] % (i + 1) == 0:
                res += i + 1
            c[i] += 1
        return res
```

```python
answers = [1,1,2]
sol = Solution()
sol.numRabbits(answers)
```
