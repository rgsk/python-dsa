```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right

import math


def isLeaf(node: Optional[TreeNode]):
    return node and node.left is None and node.right is None


def getLeafs(root: Optional[TreeNode]):
    ans: List[TreeNode] = []

    def helper(node: Optional[TreeNode]):
        if node:
            if isLeaf(node):
                ans.append(node)
            else:
                helper(node.left)
                helper(node.right)
    helper(root)
    return ans

# accepted
class Solution:
    def isValidBST(self, root: Optional[TreeNode]) -> bool:
        def helper(root: Optional[TreeNode], lower: int, upper: int):
            if root:
                if root.val <= lower or root.val >= upper:
                    return False
                return helper(root.left, lower, root.val) and helper(root.right, root.val, upper)
            return True
        return helper(root, -math.inf, math.inf)
    
    def canMerge(self, trees: List[TreeNode]) -> Optional[TreeNode]:
        toBeMerged = {tree.val: tree for tree in trees}
        all_leaves = set()
        for i in range(len(trees)):
            leafs = getLeafs(trees[i].left) + getLeafs(trees[i].right)
            for leaf in leafs:
                all_leaves.add(leaf.val)
        # if root is not the leaf of other trees
        # it means that can't be attached to other trees
        # it means this could be our final tree
        # if there is one such root
        # if leaves of this root consume all other trees
        # then this will be our answer
        # otherwise not

        # if there is more than one such root
        # then we can start with first
        # it won't be able to consume that other trees (which can't be attached to leaf)
        
        root = None
        for tree in trees:
            if tree.val not in all_leaves:
                root = tree
                break

        if root is None:
            return None

        def helper(root):
            if root:
                if isLeaf(root) and root.val in toBeMerged:
                    root.left = toBeMerged[root.val].left
                    root.right = toBeMerged[root.val].right
                    toBeMerged.pop(root.val)
                helper(root.left)
                helper(root.right)

        toBeMerged.pop(root.val)

        helper(root)

        if self.isValidBST(root) and len(toBeMerged) == 0:
            return root
        return None
```
