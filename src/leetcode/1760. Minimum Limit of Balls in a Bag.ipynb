```python
from math import ceil
from typing import List


class Solution:
    def minimumSize(self, nums: List[int], maxOperations: int) -> int:
        def can_make(max_balls):
            ops = 0
            for num in nums:
                buckets = ceil(num / max_balls)
                splits = buckets - 1
                ops += splits
            return ops <= maxOperations

        left, right = 1, max(nums)

        while left < right:
            mid = (left + right) // 2

            if can_make(mid):
                right = mid
            else:
                left = mid + 1

        return left
```

```python
nums = [9] 
maxOperations = 2
print(Solution().minimumSize(nums, maxOperations))
nums = [2,4,8,2]
maxOperations = 4
print(Solution().minimumSize(nums, maxOperations))
```

```python
class Solution:
    def minimumSize(self, nums: List[int], maxOperations: int) -> int:
        def can_make(max_balls):
            ops = 0
            for num in nums:
                ops += (num - 1) // max_balls
            return ops <= maxOperations

        left, right = 1, max(nums)

        while left < right:
            mid = (left + right) // 2

            if can_make(mid):
                right = mid
            else:
                left = mid + 1

        return left
```
