```python
import math
from typing import List

# O(n^2)
class Solution:
    def maxRotateFunction(self, nums: List[int]) -> int:
        n = len(nums)
        ans = -math.inf
        for rotate in range(n):
            s = sum([((i + rotate)%n) * nums[i] for i in range(n)])
            ans = max(ans, s)
        return ans
```

```python

class Solution:
    def maxRotateFunction(self, nums: List[int]) -> int:
        n = len(nums)
        total = 0
        current = 0
        for i in range(n):
            total += nums[i]
            current += i * nums[i]
        ans = current
        for pivot in range(n-1, 0, -1):
            current = current + total - (n * nums[pivot])
            ans = max(ans, current)
        return ans
```

```python
nums = [4,3,2,6]
sol = Solution()
sol.maxRotateFunction(nums)
```
