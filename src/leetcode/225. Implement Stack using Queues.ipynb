```python
# this performs better as push operation is O(n)
# both pop and top are O(1)

class MyStack:

    def __init__(self):
        self.q = Queue()

    def push(self, x: int) -> None:
        tempq = self.q
        self.q = Queue()
        self.q.enqueue(x)
        while tempq.size:
            self.q.enqueue(tempq.dequeue())

    def pop(self) -> int:
        return self.q.dequeue()

    def top(self) -> int:
        return self.q.front()

    def empty(self) -> bool:
        return self.q.is_empty()


# Your MyStack object will be instantiated and called as such:
# obj = MyStack()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.top()
# param_4 = obj.empty()
```

```python
class MyStack:

    def __init__(self):
        self.q1 = Queue()
        self.q2 = Queue()

    def push(self, x: int) -> None:
        self.q2.enqueue(x)

    def transfer(self):
        q1_copy = self.q1
        self.q1 = Queue()
        self.transfer_q2_to_q1()
        for _ in range(q1_copy.size):
            self.q1.enqueue(q1_copy.dequeue())
        
    def transfer_q2_to_q1(self):
        if not self.q2.is_empty():
            elem = self.q2.dequeue()
            self.transfer_q2_to_q1()
            self.q1.enqueue(elem)

    def pop(self) -> int:
        self.transfer()
        return self.q1.dequeue()

    def top(self) -> int:
        self.transfer()
        return self.q1.front()

    def empty(self) -> bool:
        return self.q1.is_empty() and self.q2.is_empty()


# Your MyStack object will be instantiated and called as such:
# obj = MyStack()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.top()
# param_4 = obj.empty()
```

```python
class MyStack:

    def __init__(self):
        self.q1 = Queue()
        self.q2 = Queue()

    def push(self, x: int) -> None:
        self.q2.enqueue(x)

    def transfer(self, frm: Queue, to: Queue):
        for _ in range(frm.size):
            to.enqueue(frm.dequeue())
        
    def transfer_reverse(self, frm: Queue, to: Queue):
        if frm.size:
            elm = frm.dequeue()
            self.transfer_reverse(frm, to)
            to.enqueue(elm)

    def perform_shifting(self):
        q1_copy = self.q1
        self.q1 = Queue()
        self.transfer_reverse(self.q2, self.q1)
        self.transfer(q1_copy, self.q1)

    def pop(self) -> int:
        self.perform_shifting()
        return self.q1.dequeue()

    def top(self) -> int:
        self.perform_shifting()
        return self.q1.front()

    def empty(self) -> bool:
        return self.q1.is_empty() and self.q2.is_empty()


# Your MyStack object will be instantiated and called as such:
# obj = MyStack()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.top()
# param_4 = obj.empty()
```
