```python
from typing import List


class Solution:
    def numMovesStonesII(self, stones: List[int]) -> List[int]:
        stones.sort()
        n = len(stones)

        # Calculate the maximum number of moves
        max_moves = stones[n-1] - stones[0] + 1 - n
        max_moves -= min(stones[1] - stones[0] - 1, stones[n-1] - stones[n-2] - 1)

        # Calculate the minimum number of moves using sliding window
        min_moves = float('inf')
        i = 0
        for j in range(n):
            while stones[j] - stones[i] + 1 > n:
                i += 1
            current_window = j - i + 1
            if current_window == n - 1 and stones[j] - stones[i] + 1 == n - 1:
                min_moves = min(min_moves, 2)
            else:
                min_moves = min(min_moves, n - current_window)

        return [min_moves, max_moves]
```

```python
from typing import List


class Solution:
    def numMovesStonesII(self, stones: List[int]) -> List[int]:
        stones.sort()
        n = len(stones)

        # Calculate the maximum number of moves
        max_moves = max(
            (stones[-1] - stones[1] + 1) - (n-1),
            (stones[-2] - stones[0] + 1) - (n-1)
        )

        # Calculate the minimum number of moves using sliding window
        min_moves = float('inf')
        i = 0
        for j in range(n):
            while stones[j] - stones[i] + 1 > n:
                i += 1
            current_window = j - i + 1
            if current_window == n - 1 and stones[j] - stones[i] + 1 == n - 1:
                min_moves = min(min_moves, 2)
            else:
                min_moves = min(min_moves, n - current_window)

        return [min_moves, max_moves]
```
