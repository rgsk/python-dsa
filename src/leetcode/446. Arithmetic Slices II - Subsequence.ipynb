```python
from collections import defaultdict
from typing import List


class Solution:
    def numberOfArithmeticSlices(self, nums: List[int]) -> int:
        n = len(nums)
        mp = {}
        for i in range(n):
            mp[i] = defaultdict(int)
        ans = 0
        for i in range(1, n):
            for j in range(i):
                diff = nums[i] - nums[j]
                mp[i][diff] += 1 + mp[j][diff]
                ans += mp[j][diff]
        return ans
```

```python
nums = [2,4,6,8,10]
sol = Solution()
sol.numberOfArithmeticSlices(nums)
```
