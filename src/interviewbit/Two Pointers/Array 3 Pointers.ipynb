```python
import math

def lower_bound(arr, target):
    low, high = 0, len(arr) - 1
    closest_index = None

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == target:
            return mid  # Target found, return its index
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

        # Update the closest index
        if closest_index is None or abs(arr[mid] - target) < abs(arr[closest_index] - target):
            closest_index = mid

    return closest_index


# TLE
class Solution:
    # @param A : tuple of integers
    # @param B : tuple of integers
    # @param C : tuple of integers
    # @return an integer
    def minimize(self, A, B, C):
        ans = math.inf
        for i in range(len(A)):
            pB = lower_bound(B, A[i])
            pC = lower_bound(C, A[i])
            maxOf3 = max(A[i], B[pB], C[pC])
            minOf3 = min(A[i], B[pB], C[pC])
            ans = min(ans, maxOf3 - minOf3)
            pB = lower_bound(B, A[i])
            pC = lower_bound(C, B[pB])
            maxOf3 = max(A[i], B[pB], C[pC])
            minOf3 = min(A[i], B[pB], C[pC])
            ans = min(ans, maxOf3 - minOf3)
            pC = lower_bound(C, A[i])
            pB = lower_bound(B, C[pC])
            maxOf3 = max(A[i], B[pB], C[pC])
            minOf3 = min(A[i], B[pB], C[pC])
            ans = min(ans, maxOf3 - minOf3)
        return ans
```
