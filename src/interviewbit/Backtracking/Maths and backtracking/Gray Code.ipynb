```python
class Solution:
    # @param A : integer
    # @return a list of integers
    def grayCode(self, A):
        def helper(n):
            if n == 1:
                return ['0', '1']
            last = helper(n - 1)
            cur = []
            for i in range(len(last)):
                cur.append('0' + last[i])
            for i in range(len(last)-1, -1, -1):
                cur.append('1' + last[i])
            return cur
        codes = helper(A)
        nums = [int(v, 2) for v in codes]
        return nums
```

```python
A = 2
sol = Solution()
sol.grayCode(A)
```
