```python
# https://www.interviewbit.com/problems/word-break-ii/
```

```python
class Solution:
    # @param A : string
    # @param B : list of strings
    # @return a list of strings
    def wordBreak(self, A, B):
        ans = []
        B = set(B)
        B = list(B)
        B.sort()
        def helper(s: str, cur):
            if s == '':
                ans.append(" ".join(cur))
                return
            for word in B:
                if s.startswith(word):
                    cur.append(word)
                    helper(s[len(word):], cur)
                    cur.pop()
        helper(A, [])
        return ans
```

```python
A = "catsanddog"
B = ["cat", "cats", "and", "sand", "dog"]
sol = Solution()
sol.wordBreak(A, B)
```
