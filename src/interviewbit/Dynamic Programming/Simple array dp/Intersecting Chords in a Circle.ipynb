```python
mod = 10**9 + 7

# little simple approach
class Solution:
    # @param A : integer
    # @return an integer
    def chordCnt(self, A):
        dp = [0 for _ in range(A + 1)]
        dp[0] = 1
        dp[1] = 1
        for i in range(2, A + 1):
            ans = 0
            l = 0
            r = i - 1
            while l <= i-1:
                ans += dp[l] * dp[r]
                l += 1
                r -= 1
            dp[i] = ans
        return dp[A] % mod
```

```python
A = 1 # 2 points
sol = Solution()
sol.chordCnt(A) # 1
```
