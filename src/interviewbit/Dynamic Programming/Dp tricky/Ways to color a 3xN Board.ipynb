```python
mod = 10**9 + 7
class Solution:
    # @param A : integer
    # @return an integer
    def solve(self, A):
        col3=24
        col2=12
        num=col2+col3
        for i in range(2,A+1):
            col3,col2=(11*col3 + 10*col2),(5*col3 + 7*col2)
            num=(col2+col3)
        return num % mod
```

```python
A = 1
sol = Solution()
sol.solve(A)
```
