```python
import math

# best solution if we store vertical sums
# Space O(1)
# Time O(n^2)
# 4969 ms
class Solution:
    # @param A : list of list of integers
    # @param B : integer
    # @return an integer
    def solve(self, A, B):
        n = len(A)
        ans = -math.inf
        for i in range(1, n):
            for j in range(n):
                A[i][j] += A[i - 1][j]

        for i in range(n - B + 1):
            s = 0
            for j in range(B):
                s += A[i+B-1][j] - (A[i-1][j] if i > 0 else 0)

            ans = max(ans, s)
            for j in range(B, n):
                s -= A[i+B-1][j-B] - (A[i-1][j-B] if i > 0 else 0)
                s += A[i+B-1][j] - (A[i-1][j] if i > 0 else 0)
                ans = max(ans, s)
        return ans
```
