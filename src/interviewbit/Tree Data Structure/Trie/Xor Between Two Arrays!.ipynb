```python
# Brute Force

# O(n^2)

class Solution:
    # @param A : list of integers
    # @param B : list of integers
    # @return an integer
    def solve(self, A, B):
        max_xor = float('-inf')
        for x in A:
            for y in B:
                max_xor = max(max_xor, x ^ y)

        return max_xor
```

```python
# amazing solution

import math


class Solution:
    # @param A : list of integers
    # @param B : list of integers
    # @return an integer
    def solve(self, A, B):
        maxofa = max(A)
        maxofb = max(B)
        maxofa_bit = int(math.log2(maxofa))
        maxofb_bit = int(math.log2(maxofb))
        max_val = 0
        if maxofa_bit >= maxofb_bit:
            for a in A:
                if int(math.log2(a)) == maxofa_bit:
                    for b in B:
                        max_val = max(max_val, a ^ b)
        if maxofb_bit >= maxofa_bit:
            for b in B:
                if int(math.log2(b)) == maxofb_bit:
                    for a in A:
                        max_val = max(max_val, b ^ a)
        return max_val
```

```python
# amazing solution

class Solution:
    # @param A : list of integers
    # @param B : list of integers
    # @return an integer
    def solve(self, A, B):
        max_xor = 0
        mask = 0

        for i in range(31, -1, -1):
            # Update the mask by setting the i-th bit
            mask |= (1 << i)

            # Create a set to store the prefixes of the numbers in array A up to the current bit
            prefixes_set = set(num & mask for num in A)

            # Update the potential maximum XOR with the numbers in array B
            potential_max_xor = max_xor | (1 << i)

            b_set = set(num & mask for num in B)

            for prefix in prefixes_set:
                # Check if there exists a prefix in array B that can give a higher XOR
                if (prefix ^ potential_max_xor) in b_set:
                    max_xor = potential_max_xor
                    break

        return max_xor
```
