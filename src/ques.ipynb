```python
input_txt = """
3
"""
```

```python
def setup(func):
    import builtins

    def wrapper(*args, **kwargs):
        old_input = builtins.input
        try:
            _it = iter(input_txt.split())
            builtins.input = lambda: next(_it)
            return func(*args, **kwargs)
        finally:
            builtins.input = old_input

    return wrapper
```

```python
@setup
def main():
    n = int(input())
    while n != 1:
        print(n, end=" ")
        if n % 2:
            n = n * 3 + 1
        else:
            n //= 2
    print(1)
```

```python
main()
```
