```python
class Solution:
    def maxSubsequence(self, nums: List[int], k: int) -> List[int]:
        heap = []

        for index, value in enumerate(nums):
            heappush(heap, (value, index))
            if len(heap) > k:
                heappop(heap)

        heap.sort(key=lambda item: item[1])
        return [value for value, _ in heap]
```
