```python
class Solution:
    def findMaxK(self, nums: List[int]) -> int:
        nums.sort()
        n = len(nums) - 1
        l, r = 0, n-1
        while l < r:
            if -nums[l] == nums[r]:
                return nums[r]
            elif -nums[l] > nums[r]:
                l += 1
            else:
                r -= 1

        return -1
```
