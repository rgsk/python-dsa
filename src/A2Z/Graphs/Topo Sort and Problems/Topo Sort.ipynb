Linear ordering of vertices such that if there is an edge between u & v, u appears before v in the ordering.

Here, we use the dfs to do the same, we ensure that all children are added to the stack before parent is added

![Screenshot 2023-10-18 at 11.37.20 PM.png](<attachment:Screenshot 2023-10-18 at 11.37.20 PM.png>)

```python
  
class Solution:
    #Function to return list containing vertices in Topological order.
    def topoSort(self, V, adj):
        # Write your code here
        visited = [False for _ in range(V)]
        st = []
        def dfs(node):
            visited[node] = True
            for child in adj[node]:
                if not visited[child]:
                    dfs(child)
            st.append(node)
        for node in range(V):
            if not visited[node]:
                dfs(node)
        return list(reversed(st))        
    
```

```python
'''
1
5 4
0 2
1 2
3 1
0 4
'''

V = 5
edges = [[2, 4], [2], [], [1], []]
sol = Solution()
sol.topoSort(V, edges)
```
