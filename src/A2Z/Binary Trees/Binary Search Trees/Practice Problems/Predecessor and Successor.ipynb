```python
from typing import List

def binary_search(nums: List[int], start: int, end: int, target: int):
    # start and end are inclusive 
    while start <= end:
        mid = (start + end)//2
        if nums[mid] == target:
            return mid
        elif target > nums[mid]:
            start = mid + 1
        else: # target < nums[mid]
            end = mid - 1
    return start


# perform the inorder traversal and then return the elements 
class Solution:
    def findPreSuc(self, root, pre, suc, key):
        # Your code goes here
        values = []
        def get_inorder(root):
            if root:
                get_inorder(root.left)
                values.append(root.key)
                get_inorder(root.right)
        get_inorder(root)

        ind = binary_search(values, 0, len(values) - 1, key)

        successor = -1
        predecessor = -1

        # we get either the index of target
        # or 1 more than target (if target doesn't exists in the array)
        if ind < len(values):
            if values[ind] > key:
                successor = ind
            else:
                if ind + 1 < len(values):
                    successor = ind + 1

        if ind - 1 >= 0 and ind - 1 < len(values):
            predecessor = ind - 1

        pre.key = values[predecessor] if predecessor != -1 else -1
        suc.key = values[successor] if successor != -1 else -1
```

```python
# while performing the inorder 
# we check if we got a better predecessor or successor
class Solution:
    def findPreSuc(self, root, pre, suc, key):
        # Your code goes here
        successor = -1
        predecessor = -1

        # do any traversal
        def perform_inorder(root):
            nonlocal predecessor, successor
            if root:
                perform_inorder(root.left)
                if root.key < key and (predecessor == -1 or root.key > predecessor):
                    predecessor = root.key
                
                elif root.key > key and (successor == -1 or root.key < successor):
                    successor = root.key

                perform_inorder(root.right)
        perform_inorder(root)

        pre.key = predecessor
        suc.key = successor
```
